## while 循环
- 当一个条件保持真的时候while循环重复执行语句
- while 循环一定要有结束条件，否则很容易进入死循环
- while 循环的语法是：

    while loop-contunuation-conndition:

        Statement

```python
i = 0
while i < 10:   #执行条件
    print('haha')
    i += 1 #代码条件（i = i + 1）
```

## for 循环
- Python的for 循环通过一个序列中的每个值来进行迭代

- b: end
- k: step
- 注意for 是循环一切可迭代对象，而不是只能使用range

## 嵌套循环
- 一个循环可以嵌套另一个循环
- 每次循环外层时，内层循环都会被刷新重新完成循环
- 也就是说，大循环执行一次，小循环会全部执行一次
- 注意：
> - 多层循环非常耗时
  - 最多使用3层循环

## 关键字 break 和 continue
- break 跳出循环，终止循环
- continue 跳出此次循环，继续执行

```python
for i in range(10):
    if i == 5:
        continue
    print(i)
```

```python
for i in range(10):
    if i == 5:
        pass
    else:
        print(i)
```

```python
for i in range(10):
    for j in range(10):
        for k in range(10):
            if k == 5:
                continue
            print(i,j,k)
```
